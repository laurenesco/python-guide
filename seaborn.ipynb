```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

basic functions:

- **sns.load_dataset(name)**: Loads a built-in dataset by name.
- **sns.set_style(style)**: Sets the style of the plots (e.g., "whitegrid", "darkgrid").
- **sns.set_palette(palette)**: Sets the color palette for the plots.
- **sns.color_palette(palette)**: Retrieves a list of colors for a specified palette.

other useful plot types:

- **sns.kdeplot(data, x)**: Creates a Kernel Density Estimate (KDE) plot.
- **sns.catplot(data, x, y, kind)**: Creates categorical plots (e.g., bar, box, violin) with a FacetGrid interface.
- **sns.violinplot(data, x, y)**: Creates a violin plot showing the distribution of a variable.
- **sns.regplot(data, x, y)**: Creates a scatter plot with a regression line.
- **sns.lmplot(data, x, y, col, row)**: Creates a scatter plot with a regression line, with faceting options.
- **sns.jointplot(data, x, y, kind)**: Creates a scatter plot with marginal histograms or KDE plots.

## scatterplot()

Usage: Creates a scatter plot of two variables. 

Common Options:

    data: DataFrame or array-like, the data to plot.
    x, y: Names of the variables to plot.
    hue: Variable for color encoding.
    style: Variable for point style.
    palette: Colors to use.

Documentation: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

Code Snippet:

```python
data = sns.load_dataset('iris')
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species', style='species')
plt.show()
```

## lineplot() 

Usage: Creates a line plot of a variable.

Common Options:

    data: DataFrame or array-like, the data to plot.
    x, y: Names of the variables to plot.
    hue: Variable for color encoding.
    style: Variable for line style.
    ci: Confidence interval to plot.

Documentation: https://seaborn.pydata.org/generated/seaborn.lineplot.html

Code Snippet:

```python
data = sns.load_dataset('flights')
sns.lineplot(data=data, x='month', y='passengers', hue='year', marker='o')
plt.show()
```

## histplot()

Usage: Creates a histogram of a variable. 

Common Options:

    data: DataFrame or array-like, the data to plot.
    x, y: Names of the variables to plot.
    bins: Number of bins.
    kde: Whether to add a KDE (Kernel Density Estimate) plot.

Documentation: https://seaborn.pydata.org/generated/seaborn.histplot.html

Code Snippet:

```python
data = sns.load_dataset('penguins')
sns.histplot(data=data, x='flipper_length_mm', bins=20, kde=True)
plt.show()
```

## boxplot()

Usage: Creates a box plot of a variable. 

Common Options:

    data: DataFrame or array-like, the data to plot.
    x, y: Names of the variables to plot.
    hue: Variable for color encoding.
    palette: Colors to use.

Code Snippet:

```python
data = sns.load_dataset('tips')
sns.boxplot(data=data, x='day', y='total_bill', hue='time')
plt.show()
```

## heatmap()

Usage: Creates a heatmap from a matrix of values. 

Common Options:

    data: 2D array-like, the data to plot.
    cmap: Colormap to use.
    annot: Whether to annotate cells with values.
    fmt: String formatting code for annotations.

Documentation: https://seaborn.pydata.org/generated/seaborn.heatmap.html

Code Snippet:

```python
data = sns.load_dataset('flights')
pivoted_data = data.pivot(index='month', columns='year', values='passengers')
sns.heatmap(pivoted_data, cmap='YlGnBu', annot=True, fmt='d')
plt.show()

# Why was this data pivoted?
#    - Original Data: A DataFrame with columns for month, year, and passengers.
#    - Pivoted Data: A matrix with month as rows, year as columns, and passengers as the values in the matrix.
```

## pairplot()

Usage: Creates pairwise plots of all variables in a DataFrame. Common Options:

    data: DataFrame or array-like, the data to plot.
    hue: Variable for color encoding.
    palette: Colors to use.
    diag_kind: Type of plot for the diagonal ("hist" or "kde").

Documentation: https://seaborn.pydata.org/generated/seaborn.pairplot.html

Code Snippet:

```python
data = sns.load_dataset('iris')
sns.pairplot(data, hue='species', diag_kind='kde')
plt.show()
```
